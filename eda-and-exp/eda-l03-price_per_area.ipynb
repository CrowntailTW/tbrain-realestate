## Price per area 

```python
df_train['log_price'] = np.log1p(df_train['total_price'])
```

```python
df_train['price_per_area'] = df_train['total_price'] / df_train['building_area']
```

```python
df_train['log_price_per_area'] = np.log1p(df_train['price_per_area'])
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_train['price_per_area'], bins=100)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df_train['log_price_per_area'], bins=100)
plt.show()
```
