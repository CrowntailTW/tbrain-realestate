### Setup

Apply Kmeans clustering to convert lat/lon 2 feats to groups and use group id as cat features.

Kmeans making script: eda-s01-kmeans-latlong.ipynb
Kmeans obj: kmeans obj saved in object_kmeans_k6_combined.pkl
K = 6

'lat', 'lon' removed from cols_num and cols_feats
'encoded_latlon' added as the group id in cols_cat and cols_feats

### Process log file

```python
saved = []
with open('log_s01_latlong-kmeans', 'r') as IFILE:
    for line in IFILE:
        if '[[((' in line:
            saved.append(line.strip())
```

```python
idx= 0
ird = 1
for i in range(2,len(saved[ird])):
    if saved[ird][i-2:i] == ']]' or saved[ird][i-2:i] == '),':
        line = saved[ird][idx:i-1].replace('[','').replace(']','')
        print(line, '  ')
        idx = i+1
```
