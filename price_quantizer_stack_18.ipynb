```python
import pandas as pd
import numpy as np
from utilities import *
```

```python
train = pd.read_csv('dataset/train.csv')
```

```python
file = 'stack_opt_18_1-10,12-22,24-29'
```

```python
test_pred = pd.read_csv('output/'+file+'.csv')
```

```python
unique_prices = np.sort(train['total_price'].unique())
```

```python
unique_prices.shape
```

```python
def find_nearest(array, value):
    return array[(np.fabs(array - value)).argmin()]

def correct_prices(df, col):
    return [find_nearest(unique_prices, x) for x in df[col]]
```

```python
test_pred['total_price'].nunique()
```

```python
test_pred['total_price_corr'] = correct_prices(test_pred, 'total_price')
```

```python
test_pred['total_price_corr'].nunique()
```
