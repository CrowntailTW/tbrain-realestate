```python
import matplotlib.pyplot as plt
```

```python
def sigmoid_01(x):
    return 1 / (1 + np.exp(-50*(x-0.1))) + 0.1*x
```

```python
x = np.arange(0, 1, 0.001)
y = sigmoid_01(x)
plt.plot(x, y); plt.xlabel('mape'); plt.grid()
plt.show()
```

```python
plt.xlim(0, 0.5)
plt.plot(x, x, label='mape')
plt.plot(x, y, label='smooth hit rate'); 
plt.plot(x, [i > 0.1 for i in x], label='hit rate', alpha=0.7)
plt.xlabel('MAPE'); plt.ylabel('loss')
plt.grid(); plt.legend()
plt.show()
```

```python
df_pred = pd.read_csv('output/model-13-predict-keras-he_uni-cv.csv')
```

```python
df_pred.head()
```

```python
def cal_score_smooth(y_true, y_pred):
    pe = np.abs(y_pred - y_true) / y_true
    return 100*np.mean(sigmoid_01(pe))
```

```python
def sigmoid_01(x):
    return 1 / (1 + np.exp(-50*(x-0.1))) + 0.1*x
```

```python
cal_score_smooth(df_pred['total_price'], df_pred['total_price_predict'])
```

```python
df_pred['ape'] = (df_pred['total_price_predict'] - df_pred['total_price']) / df_pred['total_price']
```
